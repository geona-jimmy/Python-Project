```python
from tkinter import *
x=Tk()
x.geometry("1000x1000")
x.maxsize(250,310)
x.minsize(250,310)
x.title("calculator")
a=StringVar()
def calc(num):
    global n
    n=en1.get()
    n=n+str(num)
    a.set(n)
def eql():
    global result
    result=str(eval(n))
    a.set(result)
def clr():
    c=""
    a.set(c)

def backspace():
    global result1
    current_text=en1.get()
    result1=current_text[:-1]
    a.set(result1)

en1=Entry(x,bd=3,width=25,textvariable=a)
en1.place(x=5,y=10,height=30)

clrbtn=Button(x,text="C",bg="green",command=clr)
clrbtn.place(x=190,y=9,width=42,height=42)

ninebtn=Button(x,text="9",command=lambda:calc(9))
ninebtn.place(x=5,y=55,width=42,height=42)

eight=Button(x,text="8",command=lambda:calc(8))
eight.place(relx=0.23,y=55,width=42,height=42)

seven=Button(x,text="7",command=lambda:calc(7))
seven.place(relx=0.45,y=55,width=42,height=42)

seven=Button(x,text="+",command=lambda:calc("+"))
seven.place(relx=0.75,y=55,width=42,height=42)

seven=Button(x,text="6",command=lambda:calc(6))
seven.place(x=5,y=110,width=42,height=42)

seven=Button(x,text="5",command=lambda:calc(5))
seven.place(relx=0.23,y=110,width=42,height=42)

seven=Button(x,text="4",command=lambda:calc(4))
seven.place(relx=0.45,y=110,width=42,height=42)

seven=Button(x,text="-",command=lambda:calc("-"))
seven.place(relx=0.75,y=110,width=42,height=42)

seven=Button(x,text="3",command=lambda:calc(3))
seven.place(x=5,y=165,width=42,height=42)

seven=Button(x,text="2",command=lambda:calc(2))
seven.place(relx=0.23,y=165,width=42,height=42)

seven=Button(x,text="1",command=lambda:calc(1))
seven.place(relx=0.45,y=165,width=42,height=42)

seven=Button(x,text="*",command=lambda:calc("*"))
seven.place(relx=0.75,y=165,width=42,height=42)

seven=Button(x,text="0",command=lambda:calc(0))
seven.place(x=5,y=220,width=42,height=42)

seven=Button(x,text="%",command=lambda:calc("%"))
seven.place(relx=0.23,y=220,width=42,height=42)

seven=Button(x,text="/",command=lambda:calc("/"))
seven.place(relx=0.45,y=220,width=42,height=42)

seven=Button(x,text=".",command=lambda:calc("."))
seven.place(relx=0.75,y=220,width=42,height=42)

seven=Button(x,text="=",command=eql)
seven.place(x=5,y=270,width=42,height=40)

backspace=Button(x,text="backspace",bg="green",command=backspace)
backspace.place(relx=0.19,y=270,width=120,height=40)

x.mainloop()
```
